```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("https://gist.githubusercontent.com/tijptjik/9408623/raw/b237fa5848349a14a14e5d4107dc7897c21951f5/wine.csv" , usecols=[0,1,2])
```

```python
sns.kdeplot(df["Alcohol"])
```

```python
sns.kdeplot(df["Malic_acid"])
```

```python
colors = {1:"red",2:"blue",3:"green"}
sns.scatterplot(data=df,x="Alcohol",y="Malic_acid",hue = "Class_label",palette = colors)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x= df.drop("Class_label", axis = 1)
y= df["Class_label"]
```

```python
x_train,x_test,y_train,y_test = train_test_split (x,y,test_size = 0.3,random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns = x_train.columns)
x_test_scaled = pd.DataFrame(x_train_scaled, columns = x_test.columns)
```

```python
x_train_scaled
```

```python
fig,(a,b) = plt.subplots(ncols = 2, figsize =(10,5))
a.scatter(x_train["Alcohol"],x_train["Malic_acid"],c="red")
a.set_title("Before Normalization")

b.scatter(x_train_scaled["Alcohol"], x_train_scaled["Malic_acid"], c="blue")
b.set_title("After Normalization")
```

```python
fig, (a,b) = plt.subplots(ncols = 2,figsize = (12,5))
#BEFORE SCALING (NORMALIZATION)
sns.kdeplot(x_train["Alcohol"],ax=a)
sns.kdeplot(x_train["Malic_acid"],ax=a)

#AFTER SCALLING (NORMALIZATION)
sns.kdeplot(x_train_scaled["Alcohol"], ax=b)
sns.kdeplot(x_train_scaled["Malic_acid"],ax=b)
```

```python
fig, (a,b) = plt.subplots(ncols = 2,figsize = (12,5))
#BEFORE SCALING (NORMALIZATION)
sns.kdeplot(x_train["Alcohol"],ax=a)
a.set_title("Alcohol Distribution before scaling")

#AFTER SCALLING (NORMALIZATION)
sns.kdeplot(x_train_scaled["Alcohol"], ax=b)
b.set_title("Alcohol Distribution after scaling")
```

```python
fig, (a,b) = plt.subplots(ncols = 2,figsize = (12,5))
#BEFORE SCALING (NORMALIZATION)
a.set_title = ("Malic Acid Distribution before scaling")
sns.kdeplot(x_train["Malic_acid"],ax=a)

#AFTER SCALLING (NORMALIZATION)

sns.kdeplot(x_train_scaled["Malic_acid"],ax=b)
b.set_title = ("Malic Acid Distribution after scaling")
```
